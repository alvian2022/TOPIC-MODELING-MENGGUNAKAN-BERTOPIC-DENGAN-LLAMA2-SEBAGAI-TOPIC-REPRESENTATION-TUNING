## Coherence Score perbaikan dari Alvian sumber penulis marteen (Pencipta BERTopic)

```python
documents = pd.DataFrame({"Document": tweets,
                          "ID": range(len(tweets)),
                          "Topic": topics})
```

```python
documents
```

```python
documents.dtypes
```

```python
documents['Topic'].value_counts()
```

```python
documents_per_topic = documents.groupby(['Topic'], as_index=False).agg({'Document': ' '.join})
```

```python
documents_per_topic = documents.groupby('Topic', as_index=False).agg(Document=('Document', ' '.join))
documents_per_topic['Document_Count'] = documents.groupby('Topic').size().reset_index(name='Document_Count')['Document_Count']
```

```python
documents_per_topic
```

```python
documents_per_topic['Document_Count'].sum()
```

```python
len(documents_per_topic)
```

```python
cleaned_docs = BERTopic_model._preprocess_text(documents_per_topic.Document.values)
```

```python
# Extract vectorizer and analyzer from BERTopic
vectorizer = BERTopic_model.vectorizer_model
analyzer = vectorizer.build_analyzer()
```

```python
# Use .get_feature_names_out() if you get an error with .get_feature_names()
words = vectorizer.get_feature_names_out()
```

```python
# Extract features for Topic Coherence evaluation
tokens = [analyzer(doc) for doc in cleaned_docs]
dictionary = corpora.Dictionary(tokens)
corpus = [dictionary.doc2bow(token) for token in tokens]
```

```python
BERTopic_model._outliers
```

```python
# Extract words in each topic if they are non-empty and exist in the dictionary
topic_words = []
for topic in range(-1, len(set(topics))-1):
    words = list(zip(*BERTopic_model.get_topic(topic)))[0]
    words = [word for word in words if word in dictionary.token2id]
    topic_words.append(words)
topic_words = [words for words in topic_words if len(words) > 0]
```

```python
topic_words
```

```python
# Evaluate Coherence
coherence_model = CoherenceModel(topics=topic_words,
                                 texts=tokens,
                                 corpus=corpus,
                                 dictionary=dictionary,
                                 coherence='c_v')
coherence = coherence_model.get_coherence()
```

```python
coherence
```

```python
topic_coherence_cv = coherence_model.get_coherence_per_topic(segmented_topics=None, with_std=False, with_support=False)
```

```python
len(topic_coherence_cv)
```

```python
topic_words = [[words for words, _ in BERTopic_model.get_topic(topic)]
               for topic in range(-1, len(set(topics))-1)]
```

```python
x = len(topic_words)
y = len(topic_words[0])
z = len(topic_words[0][0])
```

```python
daftar_topik = []
daftar_kata_arr = []
for i in range(x):
    daftar_kata = ""
    for j in range(y):
        o = topic_words[i][j]
        daftar_kata_arr.append(o)
    daftar_kata = ", ".join(daftar_kata_arr)
    daftar_topik.append(daftar_kata)
    daftar_kata_arr.clear()
```

```python
coherence_score_topics = pd.DataFrame(
      {
          "Topic":  range(-1, len(daftar_topik)-1),
          "Daftar_Kata": daftar_topik,
          "Cohrence Score CV": topic_coherence_cv,
      }
  )
```
